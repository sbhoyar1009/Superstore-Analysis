```python
#importing dependencies as alias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing the dataset which was downloaded from https://www.kaggle.com/jr2ngb/superstore-data
df = pd.read_csv("superstore_dataset2011-2015.csv",encoding= 'unicode_escape')
```

```python
#preview the dataset
df.head()
```

```python
#Looking at the shape and type of fields in the dataset
print("Shape of dataset is",df.shape)
df.info()
```

```python
#Check whether the data has any missing values
df.isnull().sum()
```

As we can see there are no missing values.

Since the dataset is big, there are chances of duplicate values. Check for the same, and remove it.

```python
df.duplicated().sum()
#There are no missing values
```

Now we convert all the categorical variables into numerical values by various encoding methods.

```python
sns.countplot(df["Order Priority"])
```

```python
priority_dict = {"Low":1,
                "Medium":2,
                "High":3,
                "Critical":4}
df["Order Priority"] = df["Order Priority"].map(priority_dict)
```

Since there is no missing and duplicate values, we move to descriptive analysis

```python
df.describe(include="all")
```

Exploratory Analysis

## Which Category is Best Selling and Most Profitable?

```python
category_analysis = pd.DataFrame(df.groupby(['Category'])[['Sales', 'Profit', 'Quantity']].sum())
# Set for grouped plots - figure with a 2x2 grid of Axes
sns.set_theme(style="whitegrid")
figure, axis = plt.subplots(1, 3, figsize=(8, 5))
# Plot barplots
cat1 = sns.barplot(x = category_analysis.index, y = category_analysis.Sales, ax=axis[0])
cat2 = sns.barplot(x = category_analysis.index, y = category_analysis.Profit, ax=axis[1])
cat3 = sns.barplot(x = category_analysis.index, y = category_analysis.Quantity, ax=axis[2])
# Set titles
cat1.set(title = 'Sales')
cat2.set(title = 'Profit')
cat3.set(title = 'Quantity')
# Rotate axis for x-axis
plt.setp(cat1.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat2.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat3.get_xticklabels(), rotation = 'vertical', size = 9)
# Set spacing between subplots
figure.tight_layout()
```

Do the same for the sub-category

```python
category_analysis = pd.DataFrame(df.groupby(['Sub-Category'])[['Sales', 'Profit', 'Quantity']].sum())
# Set for grouped plots - figure with a 2x2 grid of Axes
sns.set_theme(style="whitegrid")
figure, axis = plt.subplots(1, 3, figsize=(8, 5))
# Plot barplots
cat1 = sns.barplot(x = category_analysis.index, y = category_analysis.Sales, ax=axis[0])
cat2 = sns.barplot(x = category_analysis.index, y = category_analysis.Profit, ax=axis[1])
cat3 = sns.barplot(x = category_analysis.index, y = category_analysis.Quantity, ax=axis[2])
# Set titles
cat1.set(title = 'Sales')
cat2.set(title = 'Profit')
cat3.set(title = 'Quantity')
# Rotate axis for x-axis
plt.setp(cat1.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat2.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat3.get_xticklabels(), rotation = 'vertical', size = 9)
# Set spacing between subplots
figure.tight_layout()
```

## Which Customer Segment is Most Profitable?

```python
segment_analysis = pd.DataFrame(df.groupby(['Segment'])[['Profit']].sum())
# Plot Bar Plot
sns.set_theme(style="whitegrid")
sns.barplot(data = segment_analysis, x = segment_analysis.index, y = segment_analysis.Profit, palette = "rocket")
plt.title("Customer Segment Profitability")
plt.show()
```
